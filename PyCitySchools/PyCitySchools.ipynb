```python
#importing module
import pandas as pd
```

```python
#Files have been read and stored into Pandas DataFrame
school_data = pd.read_csv("Resources/schools_complete.csv")
student_data = pd.read_csv("Resources/students_complete.csv")

#Files merged ino single dataset
school_data_all = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_all.head()
```

# District Summary

```python
#Total number of unique schools in the district
school_count = len(school_data_all["school_name"].unique())
school_count 
```

```python
#Total number of students in the district
student_count = len(school_data_all["Student ID"].unique())
student_count
```

```python
#Total budget of schools in the district
schools_reduced = school_data_all.drop_duplicates("school_name") 
total_budget = schools_reduced.sum()["budget"]
total_budget
```

```python
#Average math score of students in district 
average_math_score = school_data_all["math_score"].mean()
average_math_score
```

```python
#Average reading score of students in district
average_reading_score = school_data_all["reading_score"].mean()
average_reading_score
```

```python
#Percentage of students passing math in the district
passing_math_count = school_data_all[(school_data_all["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
#Percentage of students passing reading in the district
passing_reading_count = school_data_all[(school_data_all["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
##Percentage of students passing math and reading in the district
passing_math_reading_count = school_data_all[
    (school_data_all["math_score"] >= 70) & (school_data_all["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
#Creating DF of the district
district_summary = pd.DataFrame({"Total Schools": [school_count] , "Total Students": [student_count], "Total Budget": [total_budget],
                                  "Average Math Score": [average_math_score], "Average Reading Score": [average_reading_score], 
                                   "% Passing Math": [passing_math_percentage], "% Passing Reading": [passing_reading_percentage], 
                                   "% Overall Passing Rate": [overall_passing_rate]})

#Formatting and Mapping district dataframe
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:,.3f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:,.3f}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:,.3f}".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:,.3f}".format)
district_summary["% Overall Passing Rate"] = district_summary["% Overall Passing Rate"].map("{:,.3f}".format)

#District Dataframe displayed
district_summary.head()
```

# School Summary

```python
#Setting up index
school_types = school_data.set_index(["school_name"])["type"]
```

```python
#Students Per School // 15 added in the head to include all schools
per_school_counts = school_data_all["school_name"].value_counts()
per_school_counts.head(15)
```

```python
#Total budget per school // 15 added in the head to include all schools
per_school_budget = school_data_all.groupby(["school_name"]).mean(numeric_only=True)["budget"]
per_school_budget.head(15)
```

```python
#Total budget per student // 15 added in the head to include all schools
school_budget_per_student = per_school_budget / per_school_counts
school_budget_per_student.head(15)
```

```python
# Math average score per school
per_school_math = school_data_all.groupby(["school_name"])["math_score"].mean()
per_school_math
```

```python
#Reading average score per school
per_school_reading = school_data_all.groupby(["school_name"])["reading_score"].mean()
per_school_reading
```

```python
#Percentage of students passing math per school
school_passing_math =  school_data_all.loc[(school_data_all["math_score"] >= 70)]
groupm_per_school = school_passing_math["school_name"].value_counts()
school_passing_math_perc = groupm_per_school / per_school_counts * 100
school_passing_math_perc
```

```python
#Percentage of students passing reading per school
school_passing_reading =  school_data_all.loc[(school_data_all["reading_score"] >= 70)]
groupr_per_school = school_passing_reading["school_name"].value_counts()
school_passing_reading_perc = groupr_per_school / per_school_counts * 100
school_passing_reading_perc
```

```python
#Percentage of students passing both reading and math per school
passing_math_and_reading = school_data_all[
    (school_data_all["reading_score"] >= 70) & (school_data_all["math_score"] >= 70)]
groupb_per_school = passing_math_and_reading["school_name"].value_counts()
school_passing_both_perc = groupb_per_school / per_school_counts * 100
school_passing_both_perc
```

```python
#Creating Df of school summary 
per_school_summary = pd.DataFrame({"School Type": school_types, "Total Students": per_school_counts, 
                                   "Total School Budget": per_school_budget,"Per Student Budget": school_budget_per_student, 
                                   "Average Math Score": per_school_math, "Average Reading Score": per_school_reading,
                                  "% Passing Math": school_passing_math_perc, "% Passing Reading": school_passing_reading_perc, 
                                   "% Overall Passing": school_passing_both_perc})

# Formating and Mapping school summary DF
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)
per_school_summary["Average Math Score"] = per_school_summary["Average Math Score"].map("{:,.3f}".format)
per_school_summary["Average Reading Score"] = per_school_summary["Average Reading Score"].map("{:,.3f}".format)
per_school_summary["% Passing Math"] = per_school_summary["% Passing Math"].map("{:,.3f}".format)
per_school_summary["% Passing Reading"] = per_school_summary["% Passing Reading"].map("{:,.3f}".format)
per_school_summary["% Overall Passing"] = per_school_summary["% Overall Passing"].map("{:,.3f}".format)


#Displaying school summary DF
per_school_summary.head(15)
```

# Math Scores by Grade

```python
#Seperating data by grade
ninth_graders = school_data_all[(school_data_all["grade"] == "9th")]
tenth_graders = school_data_all[(school_data_all["grade"] == "10th")]
eleventh_graders = school_data_all[(school_data_all["grade"] == "11th")]
twelfth_graders = school_data_all[(school_data_all["grade"] == "12th")]

#Assigning values to each grade
ninth_graders_scores = ninth_graders.groupby(["school_name"])["math_score"].mean()
tenth_graders_scores = tenth_graders.groupby(["school_name"])["math_score"].mean()
eleventh_graders_scores = eleventh_graders.groupby(["school_name"])["math_score"].mean()
twelfth_graders_scores = twelfth_graders.groupby(["school_name"])["math_score"].mean()

#Creating math score DF
math_score_per_grade = pd.DataFrame({"9th": ninth_graders_scores, "10th": tenth_graders_scores, 
                                    "11th": eleventh_graders_scores, "12th":twelfth_graders_scores})

#Minor data wrangling
math_score_per_grade.index.name = None

#Displaying math score DF
math_score_per_grade
```

# Reading Scores by Grade

```python
#Seperating data by grade
ninth_graders = school_data_all[(school_data_all["grade"] == "9th")]
tenth_graders = school_data_all[(school_data_all["grade"] == "10th")]
eleventh_graders = school_data_all[(school_data_all["grade"] == "11th")]
twelfth_graders = school_data_all[(school_data_all["grade"] == "12th")]

#Assigning values to each grade
ninth_graders_rscores = ninth_graders.groupby(["school_name"])["reading_score"].mean()
tenth_graders_rscores = tenth_graders.groupby(["school_name"])["reading_score"].mean()
eleventh_graders_rscores = eleventh_graders.groupby(["school_name"])["reading_score"].mean()
twelfth_graders_rscores = twelfth_graders.groupby(["school_name"])["reading_score"].mean()

#Creating reading score DF
reading_score_per_grade = pd.DataFrame({"9th": ninth_graders_rscores, "10th": tenth_graders_rscores, 
                                    "11th": eleventh_graders_rscores, "12th":twelfth_graders_rscores})

#Minor data wrangling
reading_score_per_grade.index.name = None

#Displaying reading score DF
reading_score_per_grade
```

# Score by School Spending

```python
# Establish the bins & labels
spending_bins = [0.0, 585.0, 630.0, 645.0, 680.0]

labels = ["<$585", "$585-630", "$630-645", "$645-680"]

#Copied the df to a new variable 
school_spending_df = per_school_summary.copy()

#I had some issues with the pd.cut becuase it wasnt taking the value as a float/int because of the "$" so I had to work around
#changed the value to be able to work with the bins.
school_spending_df["Per Student Budget"] = school_spending_df["Per Student Budget"].str.replace("$","", regex=False)
school_spending_df["Per Student Budget"] = school_spending_df["Per Student Budget"].astype(float)


#I had some issues also running the reast of the values since they were obejects so I just changed the typing 
school_spending_df["Average Math Score"] = school_spending_df["Average Math Score"].astype(float)
school_spending_df["Average Reading Score"] = school_spending_df["Average Reading Score"].astype(float)
school_spending_df["% Passing Math"] = school_spending_df["% Passing Math"].astype(float)
school_spending_df["% Passing Reading"] = school_spending_df["% Passing Reading"].astype(float)
school_spending_df["% Overall Passing"] = school_spending_df["% Overall Passing"].astype(float)


#"Per Student Budgets" cataegorized spending based on bins
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"], bins=spending_bins, labels=labels)
#Displayed DF
school_spending_df
```

```python
#Calculate averages for the desired columns
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean(numeric_only=True)["Average Math Score"]
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean(numeric_only=True)["Average Reading Score"]
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean(numeric_only=True)["% Passing Math"]
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean(numeric_only=True)["% Passing Reading"]
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean(numeric_only=True)["% Overall Passing"]
```

```python
#DF established
spending_summary =  pd.DataFrame({ "Average Math Score": spending_math_scores,"Average Reading Score": spending_reading_scores,
   "% Passing Math": spending_passing_math,"% Passing Reading": spending_passing_reading,
                                "% Overall Passing": overall_passing_spending,})
#DF displyed
spending_summary.head()
```

```python
# Calculate averages for the desired columns. 
size_math_scores = per_school_summary.groupby(["School Size"]).mean(numeric_only=True)["Average Math Score"]
size_reading_scores = per_school_summary.groupby(["School Size"]).mean(numeric_only=True)["Average Reading Score"]
size_passing_math = per_school_summary.groupby(["School Size"]).mean(numeric_only=True)["% Passing Math"]
size_passing_reading = per_school_summary.groupby(["School Size"]).mean(numeric_only=True)["% Passing Reading"]
size_overall_passing = per_school_summary.groupby(["School Size"]).mean(numeric_only=True)["% Overall Passing"]

#DF was created
size_summary =  pd.DataFrame({ "Average Math Score": size_math_scores,"Average Reading Score": size_reading_scores,
   "% Passing Math": size_passing_math,"% Passing Reading": size_passing_reading,
                                "% Overall Passing": size_overall_passing,})
#DF was displayed
size_summary
```
